```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Parameters
grid_sizes = [10, 20, 30]  # Example grid sizes (you can modify these)
num_graphs = 100  # Number of graphs to generate for each grid size
p = 0.5  # Probability of edge creation in Erdős–Rényi model

# Generate and store the graphs
graphs = {}

for n in grid_sizes:
    graphs[n] = []
    for i in range(num_graphs):
        G = nx.erdos_renyi_graph(n, p)
        graphs[n].append(G)

# Visualize one graph from each grid size as an example
for n in grid_sizes:
    plt.figure(figsize=(6, 6))
    G = graphs[n][0]  # Get the first graph from the list for this grid size
    nx.draw(G, with_labels=True, node_color="skyblue", node_size=500, edge_color="gray")
    plt.title(f"Erdős–Rényi Graph (n={n}, p={p})")
    plt.show()

# Now, you have 100 Erdős–Rényi graphs for each grid size stored in the 'graphs' dictionary.
```
