```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create a graph object
G = nx.Graph()

# Add the nodes (labeled from 0 to 7)
nodes = range(0, 8)
G.add_nodes_from(nodes)

# Add the edges (adjusted to match the new labels)
edges = [
    (0, 1), (1, 2), (0, 3), (3, 5), (5, 6), (6, 7), (4, 7), (2, 4)
]
G.add_edges_from(edges)

# Draw the graph
pos = {0: (0, 2), 1: (1, 2), 2: (2, 2), 3: (0, 1), 4: (2, 1), 5: (0, 0), 6: (1, 0), 7: (2, 0)}
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=500, edge_color="gray")
plt.title("Graph Representation (Nodes Labeled 0 to 7)")
plt.show()
```
